```python
import qianfan

def gen_wenxin_messages(prompt):
    '''
    构造模型请求参数 messages

    请求参数：
        prompt: 对应的用户提示词
    '''
    messages = [{"role": "user", "content": prompt}]
    return messages


def get_completion(prompt, model="ERNIE-Bot", temperature=0.01):
    '''
    获取模型调用结果

    请求参数：
        prompt: 对应的提示词
        model: 调用的模型，默认为 ERNIE-Bot，也可以按需选择 ERNIE-Bot-4 等其他模型
        temperature: 模型输出的温度系数，控制输出的随机程度，取值范围是 0~1.0，且不能设置为 0。温度系数越低，输出内容越一致。
    '''

    chat_comp = qianfan.ChatCompletion()
    message = gen_wenxin_messages(prompt)

    resp = chat_comp.do(messages=message, 
                        model=model,
                        temperature = temperature,
                        system="你是一名异常话单分析员.你已经使用一些机器学习方法分析了数据，主要发现：1. PE12的错误原因是：“特殊通话类型为 ['8163','8169']and 通话时长<60”，而特殊通话类型 主要由对方电话号码的后三位 和 总免费额度为vpmn来区分。如果对方电话号码后三位是163且是vpmn，则为8163；对方电话号码后三位是169且是vpmn，则为8169；如果对方电话号码不是163也不是169也不是vpmn，则特殊通话类型为901JJJ。  2. 如果是RN78错误原因，其直接原因是动态漫游号 match ['13900','8613900','1374','861374','1344','861344']，但因为这个错误类型的数据没有动态漫游号及对方电话号码特征。通过使用随机森林模型建模，区分RN78及PE12，发现RN78类错误，主要影响的特征是漫游费用和长途通话费用不为0，电话号码的访问地和归属地不同，且通话时长分布范围更广，相比PE12通话时间更长，对方电话归属地比较分散。")

    return resp["result"]
```
